```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
historical_data = pd.read_csv("historical_data.csv")
fear_greed_index = pd.read_csv("fear_greed_index.csv")
```

```python
historical_data
```

```python
fear_greed_index
```

```python
# Convert historical timestamp (milliseconds) and extract date
historical_data['Datetime'] = pd.to_datetime(historical_data['Timestamp'], unit='ms')
historical_data['Date'] = historical_data['Datetime'].dt.date

# Convert sentiment timestamp (seconds) and extract date
fear_greed_index['Datetime'] = pd.to_datetime(fear_greed_index['timestamp'], unit='s')
fear_greed_index['Date'] = fear_greed_index['Datetime'].dt.date
```

# Merge datasets on Date

```python
merged = pd.merge(
    historical_data,
    fear_greed_index[['Date', 'value', 'classification']],
    on='Date',
    how='left'
)
```

# Group by sentiment and analyze trader performance

```python
grouped = merged.groupby('classification').agg({
    'Closed PnL': ['mean', 'sum'],
    'Size USD': ['mean', 'sum'],
    'Account': 'nunique'
}).reset_index()

grouped.columns = ['Sentiment', 'Avg PnL', 'Total PnL', 'Avg Trade Size USD', 'Total Trade Size USD', 'Unique Traders']
grouped
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=grouped, x='Sentiment', y='Avg PnL', palette='coolwarm')
plt.title('Average PnL by Market Sentiment')
plt.ylabel('Average PnL')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Highest Profitability During "Greed":**Traders perform best during "Greed" sentiment, achieving the highest average PnL.

Suggests traders may thrive in optimistic market conditions where opportunities are more aggressively pursued.

**Moderate Success in "Fear":** Surprisingly, performance during "Fear" sentiment is better than during "Extreme Greed".

Indicates some traders might adopt more cautious, strategic behavior under fearful conditions, leading to effective trades.

**Underperformance in "Extreme Greed" and "Neutral":** "Extreme Greed" has lower average PnL than "Greed", implying traders might take on excessive risk and face losses.

"Neutral" sentiment correlates with lowest performance, possibly due to lack of strong signals leading to indecision or low-quality trades.

***Strategic Implications:***

1. Capitalize on Greed: Traders should identify "Greed" phases using sentiment analysis and consider deploying more capital or strategies proven to perform well in bullish conditions.

2. Caution During Extreme Greed: Implement stricter risk management to counteract overconfidence and excessive risk-taking.

3. Explore Defensive Strategies in Neutral Markets: Develop or enhance strategies that work in sideways/low-volatility conditions, or reduce exposure during such periods.

4. Leverage Contrarian Tactics in Fearful Markets: Consider opportunistic entries against prevailing fear if backed by data, as traders may be more conservative and leave profitable openings.

**Feature Engineering for Trader Clustering**

```python
# Aggregate key stats per trader
trader_features = merged.groupby('Account').agg({
    'Closed PnL': ['mean', 'sum'],
    'Size USD': ['mean', 'sum', 'count']
}).reset_index()

# Rename columns
trader_features.columns = [
    'Account', 'Avg PnL', 'Total PnL',
    'Avg Trade Size USD', 'Total Trade Size USD', 'Total Trades'
]

# Normalize values for clustering
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaled = scaler.fit_transform(trader_features.iloc[:, 1:])
```

**Apply KMeans Clustering to Group Traders**

```python
from sklearn.cluster import KMeans
import numpy as np

# Try 3 clusters (you can tune this)
kmeans = KMeans(n_clusters=3, random_state=42)
trader_features['Cluster'] = kmeans.fit_predict(features_scaled)

# Show sample from each cluster
for i in range(3):
    print(f"\nCluster {i} sample traders:")
    print(trader_features[trader_features['Cluster'] == i].head(3))
```

# Anomaly Detection — Detect Spikes in PnL per Trader

```python
# Calculate daily PnL for top trader
top_trader = trader_features.sort_values(by='Total PnL', ascending=False).iloc[0]['Account']
trader_daily = merged[merged['Account'] == top_trader].groupby('Date')['Closed PnL'].sum()

# Calculate z-score
from scipy.stats import zscore
trader_z = zscore(trader_daily)

# Flag anomalies (z > 2 or z < -2)
anomalies = trader_daily[np.abs(trader_z) > 2]

# Display
print(f"Anomalies for {top_trader}:")
print(anomalies)

# Plot
plt.figure(figsize=(10, 5))
trader_daily.plot(label='PnL')
anomalies.plot(style='o', color='red', label='Anomaly')
plt.title(f"Anomaly Detection for {top_trader}")
plt.legend()
plt.grid(True)
plt.show()
```

**Anomaly Detected on 2024-10-27:**

*The spike to $1.42M is flagged as statistically anomalous,
suggesting:*

1. An unusually large gain compared to the trader’s other daily returns.
2. Possibly linked to a high-impact event, large position, or external market volatility.

**Pattern Before and After:**

*Prior to this spike, PnL was growing gradually but modestly. Post-spike, performance declines steadily, indicating:*

1. The anomaly may have been non-repeatable (e.g., a rare arbitrage, insider signal, or single-day breakout).

2. The trader either scaled down or faced diminishing edge afterward.

**Contextual Link to Sentiment:**

From earlier insights, 2024-10-27 had Greed sentiment.

This suggests the trader excels during bullish windows—likely momentum or breakout-driven strategies.

# Analysis of change of Sentiments

```python
# Ensure 'Date' is sorted
# Create the sentiment_df DataFrame from fear_greed_index, aggregating by date
sentiment_df = fear_greed_index.groupby('Date')['value'].mean().reset_index()

sentiment_df = sentiment_df.sort_values(by='Date')

# Calculate daily sentiment change (absolute and signed)
sentiment_df['Sentiment_Change'] = sentiment_df['value'].diff()
sentiment_df['Sentiment_Shift'] = sentiment_df['Sentiment_Change'].abs()

# Preview
sentiment_df[['Date', 'value', 'Sentiment_Change', 'Sentiment_Shift']].head()
```

```python
# Aggregate daily PnL for all traders
daily_pnl = merged.groupby('Date')['Closed PnL'].sum().reset_index()

# Merge with sentiment shift data
sentiment_pnl = pd.merge(sentiment_df, daily_pnl, on='Date', how='inner')

# Preview
sentiment_pnl[['Date', 'Sentiment_Shift', 'Closed PnL']].head()
```

**Visualize Relationship Between Sentiment Shift and PnL**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sentiment_pnl, x='Sentiment_Shift', y='Closed PnL')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Sentiment Shift vs Total Daily PnL')
plt.xlabel('Sentiment Shift (absolute change in index)')
plt.ylabel('Total Closed PnL')
plt.grid(True)
plt.show()
```
